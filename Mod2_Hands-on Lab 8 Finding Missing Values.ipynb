### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
def most_frequent_value(col):
    vals = df.groupby(col)[col].value_counts()
    most_freq = vals.sort_values(ascending=False).index[0]
    print('Column', vals)
    nas = df[col].isna().sum()
    print('Most frequent value:', most_freq)
    print('Missing values:', nas)
    print('Total:', sum(vals)+nas)
    plt.figure(figsize=(10, 6))
    sns.barplot(vals)
    plt.xticks(rotation=90)
    plt.title(f'Distribution of {col}')
    plt.show()
    return most_freq

#most_frequent_value('MainBranch')
#most_frequent_value('Employment')
#most_frequent_value('RemoteWork')
most_frequent_value('EdLevel')
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
# no missing values in column Employment
col = 'EdLevel'
df_EdFilled = df.fillna(value={col: most_frequent_value(col)})
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
## Write your code here
plt.figure(figsize=(10, 6))
sns.barplot(df_EdFilled.groupby(col)[col].value_counts())
plt.xticks(rotation=90)
plt.title(f'Distribution of {col} after filling NA with Most Frequent Value')
```
